```python
import requests
import pandas as pd
from pandas import json_normalize
```

```python
txt_file = open("items.txt", "r")
lines =txt_file.read().split(',')
```

```python
def amazonPriceRequest(asin):
    url = "https://amazon-price-history.p.rapidapi.com/api/us/price_history"

    querystring = {"asin":asin,"price_type":"amazon"}

    headers = {
        'x-rapidapi-key': "7a03787ae6msh833680465f5a2aep1dc51cjsnbd77cf857f41",
        'x-rapidapi-host': "amazon-price-history.p.rapidapi.com"
        }

    response = requests.request("GET", url, headers=headers, params=querystring).json()
    return response
```

```python
ko = 0

for index,asin in enumerate(lines):
    
    print("calling {} from {}".format(index,len(lines)-1))
    
    try:
        json_response = amazonPriceRequest(asin)
           
        json_normalized = json_normalize(data=json_response, record_path='price_history',meta=['asin','currency','price_type'])
        if index == 0:
            json_df = json_normalized
            print("... OK")
        else:
            json_df = json_df.append(json_normalized, ignore_index=True, sort=False)
            print("... OK")   
        
    except:    
        print("ko --> %s" %(asin))
        ko+=1
print("Total number of calls %d, total number of ko %d" % (len(lines), ko))
    #json_df =json_df.append(json_normalize, ignore_index=True, sort=False)
```
